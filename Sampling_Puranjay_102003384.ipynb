# Balancing the imbalanced dataset:
## Random oversampling has been used to balance the imbalanced dataset. Random oversampling involves randomly duplicating examples from the minority class and adding them to the training dataset.

```python
from imblearn.over_sampling import RandomOverSampler
randoversam = RandomOverSampler(random_state=42)
X_r, Y_r = randoversam.fit_resample(X, Y)
```

```python
class_count_new = Y_r.value_counts(normalize=False)
print(class_count_new)
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, Y_train, Y_test = train_test_split(X_r, Y_r, test_size=0.33, random_state=42)
```

## The five models that have been used for analysis are:
1. Random forest
2. Decision tree
3. Logistic Regression
4. SVC
5. K-Nearest Neighbors 

Corresponding functions have been defined that fit the dataset into the model, make predictions and return accuracy which is then used for final comparison.

```python
from sklearn.ensemble import RandomForestClassifier
from sklearn.neighbors import KNeighborsClassifier
from sklearn.tree import DecisionTreeClassifier
from sklearn.linear_model import LogisticRegression
from sklearn.linear_model import SGDClassifier
from sklearn.svm import SVC
from sklearn.metrics import accuracy_score
```

```python
def RandomForest(X_train, Y_train, X_test, Y_test):
    clf = RandomForestClassifier(n_estimators = 2)
    clf.fit(X_train, Y_train)
    Y_pred = clf.predict(X_test)
    return accuracy_score(Y_test,Y_pred)
```

```python
def KNN(X_train, Y_train, X_test, Y_test):
    knn = KNeighborsClassifier(n_neighbors = 7).fit(X_train, Y_train)
    Y_pred = knn.predict(X_test)
    return accuracy_score(Y_test,Y_pred)
```

```python
def DecisionTree(X_train, Y_train, X_test, Y_test):
    tree = DecisionTreeClassifier(max_depth = 2).fit(X_train, Y_train)
    Y_pred = tree.predict(X_test)
    return accuracy_score(Y_test,Y_pred)
```

```python
def Logistic(X_train, Y_train, X_test, Y_test):
    clf = LogisticRegression()
    clf.fit(X_train, Y_train)
    Y_pred = clf.predict(X_test)
    return accuracy_score(Y_test,Y_pred)
```

```python
def SVC(X_train, Y_train, X_test, Y_test):
    svm = SGDClassifier(loss='hinge')
    svm.fit(X_train, Y_train)
    Y_pred = svm.predict(X_test)
    return accuracy_score(Y_test,Y_pred)
```

# The five sampling techniques have been used as follows: 
## 1. Simple random sampling

```python
import math
import random as rd
z = 1.96 #confidence level
p = 0.5
e = 0.05 #margin of error
n_srs = math.floor((z*z*p*(1-p))/(e*e))
print("n = ", n_srs)
def random_sampling(df, n_srs):
    random_sample = df.sample(n = n_srs)
    return(random_sample)

randomSample = random_sampling(df_new, n_srs)
```

```python
randomSample.head()
```

```python
X_srs = randomSample.drop('Class', axis=1) #new training X
Y_srs = randomSample['Class'] #new training Y

SRS_RandomForest = RandomForest(X_srs,Y_srs,X_test,Y_test)
SRS_DecisionTree = DecisionTree(X_srs,Y_srs,X_test,Y_test)
SRS_KNN = KNN(X_srs,Y_srs,X_test,Y_test)
SRS_Logistic = Logistic(X_srs,Y_srs,X_test,Y_test)
SRS_svc = SVC(X_srs,Y_srs,X_test,Y_test)

SRS = [SRS_RandomForest,SRS_DecisionTree,SRS_KNN,SRS_Logistic,SRS_svc]
print(SRS)
```

## 2. Stratified Sampling

```python
s = 2 #number of strata
n_strat = math.floor((z*z*p*(1-p))/(e*e/s*s))
print("n = " , n_strat)
strata = df_new.groupby('Class')

stratifiedSample = strata.apply(lambda x: x.sample(n_strat))
```

```python
stratifiedSample.head()
```

```python
X_st = stratifiedSample.drop('Class', axis=1) 
Y_st = stratifiedSample['Class'] 

ST_RandomForest = RandomForest(X_st,Y_st,X_test,Y_test)
ST_DecisionTree = DecisionTree(X_st,Y_st,X_test,Y_test)
ST_KNN = KNN(X_st,Y_st,X_test,Y_test)
ST_Logistic = Logistic(X_st,Y_st,X_test,Y_test)
ST_svc = SVC(X_st,Y_st,X_test,Y_test)

ST = [ST_RandomForest,ST_DecisionTree,ST_KNN,ST_Logistic,ST_svc]
print(ST)
```

## 3. Cluster Sampling

```python
from sklearn.cluster import KMeans
kmeans = KMeans(n_clusters=10, random_state=42).fit(df_new) #number of clusters = 10
cluster_assignments = kmeans.labels_
cluster = [1,3,5,6,8]
cluster_series = pd.Series(cluster_assignments)

cluster_df = df_new[cluster_series.isin(cluster)]
```

```python
cluster_df.head()
```

```python
X_c = cluster_df.drop('Class', axis=1) 
Y_c = cluster_df['Class'] 

C_RandomForest = RandomForest(X_c,Y_c,X_test,Y_test)
C_DecisionTree = DecisionTree(X_c,Y_c,X_test,Y_test)
C_KNN = KNN(X_c,Y_c,X_test,Y_test)
C_Logistic = Logistic(X_c,Y_c,X_test,Y_test)
C_svc = SVC(X_c,Y_c,X_test,Y_test)

C = [C_RandomForest,C_DecisionTree,C_KNN,C_Logistic,C_svc]
print(C)
```

## 4. Systematic Sampling

```python
systematic_df.head()
```

```python
X_sy = systematic_df.drop('Class', axis=1) 
Y_sy = systematic_df['Class'] 

SY_RandomForest = RandomForest(X_sy,Y_sy,X_test,Y_test)
SY_DecisionTree = DecisionTree(X_sy,Y_sy,X_test,Y_test)
SY_KNN = KNN(X_sy,Y_sy,X_test,Y_test)
SY_Logistic = Logistic(X_sy,Y_sy,X_test,Y_test)
SY_svc = SVC(X_sy,Y_sy,X_test,Y_test)

SY = [SY_RandomForest,SY_DecisionTree,SY_KNN,SY_Logistic,SY_svc]
print(SY)
```

## 5. Reservoir Sampling

```python
def sample_reservoir(dataframe, sample_size, state = None):
    stream = []
    size = len(dataframe)
    for i in range(size):
        stream.append(i)
    i = 0
    reservoir = [0] * sample_size
    for i in range(sample_size):
        reservoir[i] = stream[i]
    rd.seed(state)
    while i < size:
        j = rd.randrange(i + 1)
        if j < sample_size:
            reservoir[j] = stream[i]
        i += 1
    
    return dataframe.iloc[reservoir]
reservoir_df = sample_reservoir(df_new, 400, 42)
```

```python
reservoir_df.head()
```

```python
X_res = reservoir_df.drop('Class', axis=1)
Y_res = reservoir_df['Class']

R_RandomForest = RandomForest(X_res,Y_res,X_test,Y_test)
R_DecisionTree = DecisionTree(X_res,Y_res,X_test,Y_test)
R_KNN = KNN(X_res,Y_res,X_test,Y_test)
R_Logistic = Logistic(X_res,Y_res,X_test,Y_test)
R_svc = SVC(X_res,Y_res,X_test,Y_test)

R = [R_RandomForest,R_DecisionTree,R_KNN,R_Logistic,R_svc]
print(R)
```
